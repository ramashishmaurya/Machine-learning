```python
from sklearn.impute import SimpleImputer
model = SimpleImputer(strategy='mean')
```

```python
models = SimpleImputer(strategy='most_frequent')
```

```python
df[['deck']] = models.fit_transform(df[['deck'])
```

```python
def cleancolumn(columnname:str): # this made work easy 
    df[[columnname]] = models.fit_transform(df[[columnname]])
```
