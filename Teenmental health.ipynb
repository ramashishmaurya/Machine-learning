```python
import pandas as pd 
import numpy as np 
```

```python
df = pd.read_csv('D:/datasets/Teen_Mental_Health_Dataset.csv')
```

```python
df.sample()
```

```python
df.columns
```

```python
df['depression_label'].value_counts()
```

```python
x = df.drop('depression_label', axis=1)
```

```python
y = df[['depression_label']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train ,x_test , y_train , y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
df.isnull().sum()
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
trans = ColumnTransformer(
    transformers=[
        ('onehot', OneHotEncoder(sparse_output=False), [1, 3, 8])
    ],
    remainder='passthrough'
)
```

```python
from sklearn.pipeline import Pipeline
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pipe = Pipeline(steps=[
    ('trans' ,trans), 
    ('model' ,DecisionTreeClassifier())
    
])
```

```python
pipe.named_steps['trans']
```

```python
pipe.fit(x_train , y_train)
```

```python
y_pre = pipe.predict(x_test)
```

```python
y_pre
```

```python
from sklearn.metrics import mean_absolute_error, r2_score ,accuracy_score

print("MAE:", mean_absolute_error(y_test, y_pre))
print("R2 Score:", r2_score(y_test, y_pre))
```

```python
accuracy_score(y_test , y_pre)
```
