```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df['sex'].value_counts()
```

```python
plt.hist(df['sex'].value_counts()  )
plt.xlabel('Categories of male ')
plt.ylabel('Categories of female')
```

```python
df.sample(3)
```

```python
from sklearn.impute import SimpleImputer
```

```python
simple = SimpleImputer(strategy='mean')
```

```python
df['age'] = simple.fit_transform(df[['age']])
```

```python
df.sample(4)
```

```python
x =df[['age','fare','pclass']]
y = df[['survived']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x ,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
tree = DecisionTreeRegressor(max_depth=4)
```

```python
tree.fit(x_train , y_train)
```

```python
y_pre = tree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report , r2_score , mean_absolute_error
```

```python
r2_score(y_test , y_pre)
```

```python
mean_absolute_error(y_test , y_pre)
```
