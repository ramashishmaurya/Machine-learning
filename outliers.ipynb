```python
df.head()
```

```python
counts = df['Country'].value_counts().index
values =  df['Country'].value_counts().values
```

```python
plt.xlabel('Name of the country ')
sns.barplot(x=counts , y=values , palette=sns.color_palette("husl", len(counts)) )
plt.ylabel('Spending usd ')
```

```python
df.groupby('Country')['Life_Expectancy'].mean() 
```

```python
plt.xlabel('name of country')
plt.ylabel('Life Expectancy')
sns.barplot(x=df.groupby('Country')['Life_Expectancy'].mean().index ,
            y=df.groupby('Country')['Life_Expectancy'].mean().values , 
            palette=['red' , 'yellow' ,'pink'])
import warnings
```
